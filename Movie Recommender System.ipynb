```python
import numpy as np
import pandas as pd
#MRS = Movie recommender system
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.head()
```

```python
movies = movies.merge(credits,on='title') #merging both datasets
```

```python
movies = movies[['movie_id', 'title','overview','genres','keywords','cast','crew']]
#keeping only relevant data for MRS in movies 
```

```python
movies.head()
```

```python
movies.isnull().sum() #checking missing data
```

```python
movies.dropna(inplace=True) #dropping missing values in dataset
#inplace=true used to implement drop on same dataset and not on new dataset
movies.isnull().sum()
```

```python
movies['genres'][0] #genre column for first movie
```

```python
#helper function to convert genres with only relevant genre data
#literal_eval converts the string into a list using import ast
import ast
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
   
```

```python
movies['genres'] = movies['genres'].apply(convert)
#converting the genres into only required data(only name of genre needed)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#function to get first 3 actors name in the dataset as they are the
#relevent data for MRS
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break  
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#function to find director in all crew
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director) 
#only director name is needed
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
#converts the overview into a list containing strings seperated by commas
```

```python
#removing spaces between words so that there is no confusion
#for eg for string Sam it can recommend Sam worthington or Sam raymond 
#so remove all spaces and store as single string
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#concatenating all columns into a new column called tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast']+ movies['crew']
```

```python
movies.head()
```

```python
#creating new dataframe df with only movie_id, title and tags
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
#converting tags list into string and joining spaces
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
#converting all strings to lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#problem statement is that when a user enters a movie we have to return top 5 movies closest to the given movie
#Core idea we are going to use is converting the texts to vectors
#we will return the 5 vectors that are closest to input vector
#text to vectors are called text vectorization
#technique we will use is bag of words
#other techniques are tfidf, wordtovec etc.
#we will perform vectorization only on main words and not on stop words
#such as is and but etc.
```

```python
#sklearn library to do vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
#when we check features of count vectorizer we see repeating words
#for eg. action,actions etc.
#to remove these redundant words with same meanings we use stemming
#stemming will convert all words related to action to 'action'
#we will use nltk library to apply stemming
#!pip install nltk
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#we will find cosine distance between the vectors and not euclidean dist
#cosine distance = angle between two vectors
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]  
```

```python
#creating a function for recommendation
def recommend(movie):
  #fetching index of movie entered
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  #enumerate keyword helps sorting the distances while keeping a check 
  #of the index of each movie
  #we have to sort in descending on the basis of how similar they are
  #closer the value to 1, better the similarity as angle increases 
  #value of cos(angle) decreases
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
